```python
from datasets import load_dataset

# Load the dataset
dataset = load_dataset('devanshamin/PubMedDiabetes-LLM-Predictions')

# Shuffle the dataset and slice it
dataset = dataset['train'].shuffle(seed=42).select(range(1000))

# Define a function to transform the data
def transform_pubmed(example):
    title = example['title'].strip()
    abstract = example['abstract'].strip()
    explanation = example['explanation'].strip()

    # Apply a template similar to your previous transformation
    transformed_text = f"<s>[INST]{title}[/INST] {abstract} \n {explanation} </s>"

    return {'text': transformed_text}

# Apply the transformation
transformed_dataset = dataset.map(transform_pubmed).remove_columns([col for col in dataset.column_names if col != 'text'])

# Check a sample output
print(transformed_dataset[0])
```
